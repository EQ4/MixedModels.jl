# Boundaries of allowable correlations for k = 3

A $2\times 2$ correlation matrix is of the form
$\begin{bmatrix}1.&r\\r&1.\end{bmatrix}$
for any $r$ in [-1.,1.], which makes it easy to determine when $r=\rho_{1,2}$, is on the boundary.

For the $3\times 3$ case the relationships determining the boundary for allowable correlation coefficients are more complicated.  Suppose that $\rho_{1,2}=r$.  The allowable values of $x = \rho_{1,3}$ and $y=\rho_{2,3}$ are bounded by the ellipse in the $x,y$ plane with conjugate axes $y=rx$ and $x=ry$.  These points can be expressed as $x = \cos(\omega)$, $y=\cos(\omega+\cos^{-1}(r))$ for $\omega\in[-\pi,\pi]$.

A function to produce such a slice is

```julia
using PyPlot
const omega = linrange(-pi,pi,201)
const x = cos(omega)
function yvals(r::Number)
    -1. < r < 1. || error("r = $r is not in (-1,1)")
    cos(omega + acos(r))
end
```

```julia
plot(x,yvals(0.),color="blue");
```

```julia
plot(x,yvals(0.6),color="blue");
```

```julia
plot(x,yvals(-0.6),color="blue");
```

## Derivation

The easiest way to derive the result is through the Cholesky factor of the correlation matrix.  If the correlation matrix is of the form
$$\begin{bmatrix}
1&r&x\\
r&1&y\\
x&y&1
\end{bmatrix}$$
its lower Cholesky factor is
$$\begin{bmatrix}
1&0&0\\
r&\sqrt{1-r^2}&0\\
x&

\end{bmatrix}$$
